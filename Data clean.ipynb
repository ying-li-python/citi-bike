## Citi Bike Trip History Data clean
- Source: https://www.citibikenyc.com/system-data


```python
# import dependencies 
import pandas as pd
```

```python
# read data from csv as panda dataframe
df1 = pd.read_csv("Resources/JC-201805-citibike-tripdata.csv")
df2 = pd.read_csv("Resources/JC-201806-citibike-tripdata.csv")
df3 = pd.read_csv("Resources/JC-201807-citibike-tripdata.csv")
df4 = pd.read_csv("Resources/JC-201808-citibike-tripdata.csv")
df5 = pd.read_csv("Resources/JC-201809-citibike-tripdata.csv")
df6 = pd.read_csv("Resources/JC-201810-citibike-tripdata.csv")
df7 = pd.read_csv("Resources/JC-201811-citibike-tripdata.csv")
df8 = pd.read_csv("Resources/JC-201812-citibike-tripdata.csv")
df9 = pd.read_csv("Resources/JC-201901-citibike-tripdata.csv")
df10 = pd.read_csv("Resources/JC-201902-citibike-tripdata.csv")
df11 = pd.read_csv("Resources/JC-201903-citibike-tripdata.csv")
df12 = pd.read_csv("Resources/JC-201904-citibike-tripdata.csv")
df13 = pd.read_csv("Resources/JC-201905-citibike-tripdata.csv")

# append dataframes together
cleaned_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13])

# check dataframe
cleaned_df.head()
```
